```python
# renaimg the patient id to client id 
data.rename(columns={'Patient ID':'Client ID'} ,inplace= True)
```

```python
data.head()
```

```python
# Triming the PAT000 in the begining od the client id column 
data['Client ID'] = data['Client ID'].str.strip('PAT000')
```

```python
data.sample(10)
```

```python
# Combining first name and last name to form the full name 
data['Full Name']=data['First Name']+ "  " +   data['Last Name']
```

```python
data.sample(2)
```

```python
pd.to_datetime(data['Requesting date'])
pd.to_datetime(data['Appointment date'])
pd.Series(data['Requesting date'])
pd.Series(data['Appointment date'])
data['Waiting Time']=data['Appointment date'] - data['Requesting date']
```

```python
data.head()
```

```python
data['Waiting Time'] = pd.to_timedelta(data['Waiting Time'])
```

```python
#formatting the waiting time into days and hours 
data['Waiting_Time'] = data['Waiting Time'].dt.days.astype(str) + ' days ' + \
                               (data['Waiting Time'].dt.seconds // 3600).astype(str) + ' hours'
```

```python
data.head()
```

```python
# droping the unformated wait time 
data.drop(columns=['Waiting Time'],inplace = True)
```
